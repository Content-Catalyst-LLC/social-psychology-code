# Heuristics and Biases

This notebook scaffold loads judgment-under-uncertainty trial data and prepares summaries for anchoring error, calibration error, overconfidence, framing, availability, representativeness, affective risk judgment, and decision quality.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/heuristics_biases_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/heuristics_biases_model.py --simulate first')

df = pd.read_csv(data_path)
df['anchoring_error'] = df['estimate'] - df['true_value']
df['absolute_error'] = df['anchoring_error'].abs()
df['calibration_error'] = (df['confidence_rating'] - df['actual_accuracy']).abs()
df.head()
```

```python
summary = (
    df.groupby(['heuristic_type', 'condition'])
    .agg(
        n=('participant', 'size'),
        mean_abs_error=('absolute_error', 'mean'),
        mean_calibration_error=('calibration_error', 'mean'),
        mean_overconfidence=('overconfidence_score', 'mean'),
        mean_decision_quality=('decision_quality', 'mean')
    )
    .reset_index()
)
summary
```

# Social Comparison Theory

This notebook scaffold loads the social comparison dataset and prepares summaries for upward, downward, lateral, digital, organizational, and relative-deprivation comparison dynamics.

```python
from pathlib import Path
import pandas as pd

data_path = Path('../data/social_comparison_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/social_comparison_model.py --simulate first')

df = pd.read_csv(data_path)
df['self_eval_change'] = df['self_eval_post'] - df['self_eval_pre']
df.head()
```

```python
summary = (
    df.groupby(['condition', 'comparison_type'])
    .agg(
        n=('self_eval_change', 'size'),
        mean_gap=('comparison_gap', 'mean'),
        mean_attainability=('attainability', 'mean'),
        mean_self_eval_change=('self_eval_change', 'mean'),
        mean_motivation=('motivation_score', 'mean'),
        mean_envy=('envy', 'mean'),
        mean_inspiration=('inspiration', 'mean'),
        mean_discouragement=('discouragement', 'mean'),
        mean_relative_deprivation=('relative_deprivation', 'mean')
    )
    .reset_index()
)
summary
```

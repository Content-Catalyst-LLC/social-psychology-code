# Cognitive Dissonance Theory

This notebook scaffold loads dissonance trial data and prepares summaries for attitude change, spreading of alternatives, effort justification, belief disconfirmation, self-affirmation, response latency, and institutional escalation.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/dissonance_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/cognitive_dissonance_model.py --simulate first')

df = pd.read_csv(data_path)
df['attitude_change'] = df['post_attitude'] - df['pre_attitude']
df['spreading_of_alternatives'] = (df['chosen_post_value'] - df['chosen_pre_value']) - (df['rejected_post_value'] - df['rejected_pre_value'])
df.head()
```

```python
summary = (
    df.groupby(['paradigm', 'condition'])
    .agg(
        n=('participant', 'size'),
        mean_attitude_change=('attitude_change', 'mean'),
        mean_spreading=('spreading_of_alternatives', 'mean'),
        mean_discomfort=('dissonance_discomfort', 'mean'),
        mean_reversal=('policy_reversal_willingness', 'mean')
    )
    .reset_index()
)
summary
```

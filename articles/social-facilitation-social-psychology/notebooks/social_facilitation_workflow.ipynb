# Social Facilitation

This notebook scaffold loads social facilitation data and prepares summaries for audience presence, evaluation pressure, task difficulty, task mastery, arousal, distraction, response time, and performance.

```python
from pathlib import Path
import pandas as pd

data_path = Path('../data/social_facilitation_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/social_facilitation_model.py --simulate first')

df = pd.read_csv(data_path)
df['social_presence_intensity'] = df['audience_present'] + 0.65 * df['coaction_present'] + 0.50 * df['digital_monitoring']
df['mastery_advantage'] = df['task_mastery'] - df['task_difficulty']
df.head()
```

```python
summary = (
    df.groupby(['condition', 'task_domain'])
    .agg(
        n=('participant', 'size'),
        mean_performance=('performance_score', 'mean'),
        mean_accuracy=('accuracy', 'mean'),
        mean_arousal=('arousal_index', 'mean'),
        mean_evaluation=('evaluation_pressure', 'mean'),
        mean_response_time=('response_time_ms', 'mean')
    )
    .reset_index()
)
summary
```

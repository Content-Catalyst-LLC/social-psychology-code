# Group Polarization and Collective Judgment

This notebook scaffold loads group-polarization data and prepares summaries for attitude shift, extremity amplification, confidence shift, persuasive arguments, social comparison, identity salience, norm enforcement, dissent quality, deliberation structure, and collective judgment quality.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/group_polarization_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/group_polarization_model.py --simulate first')

df = pd.read_csv(data_path)
df['attitude_shift'] = df['post_attitude'] - df['pre_attitude']
df['extremity_shift'] = df['post_attitude'].abs() - df['pre_attitude'].abs()
df['confidence_shift'] = df['post_confidence'] - df['pre_confidence']
df.head()
```

```python
summary = (
    df.groupby(['condition', 'platform_context'])
    .agg(
        n=('participant', 'size'),
        mean_shift=('attitude_shift', 'mean'),
        mean_extremity_shift=('extremity_shift', 'mean'),
        mean_confidence_shift=('confidence_shift', 'mean'),
        mean_decision_quality=('decision_quality', 'mean'),
        mean_accuracy=('collective_judgment_accuracy', 'mean')
    )
    .reset_index()
)
summary
```

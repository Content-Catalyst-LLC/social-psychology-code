# Attribution Theory

This notebook scaffold loads attribution trial data and prepares summaries for internal attribution, external attribution, responsibility, blame, sympathy, hostile attribution, and institutional context.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/attribution_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/attribution_theory_model.py --simulate first')

df = pd.read_csv(data_path)
df['disposition_bias_index'] = df['attribution_internal'] - df['attribution_external']
df['covariation_situation_index'] = (df['consensus'] + df['distinctiveness'] + df['consistency']) / 3
df.head()
```

```python
summary = (
    df.groupby(['condition', 'target_type', 'outcome_valence'])
    .agg(
        n=('participant', 'size'),
        mean_internal=('attribution_internal', 'mean'),
        mean_external=('attribution_external', 'mean'),
        mean_responsibility=('responsibility_rating', 'mean'),
        mean_blame=('blame_rating', 'mean'),
        mean_help=('help_support', 'mean')
    )
    .reset_index()
)
summary
```

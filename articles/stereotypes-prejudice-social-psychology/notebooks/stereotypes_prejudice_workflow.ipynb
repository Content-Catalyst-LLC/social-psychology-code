# Stereotypes, Prejudice, and Discrimination

This notebook scaffold loads stereotype/prejudice trial data and prepares summaries for stereotype strength, prejudice, discrimination tendency, contact support, stereotype threat, identity safety, and institutional disparity.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/stereotypes_prejudice_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/stereotypes_prejudice_model.py --simulate first')

df = pd.read_csv(data_path)
df['contact_support_index'] = (df['contact_quality'] + df['contact_quantity'] + df['institutional_support']) / 3
df['threat_competition_index'] = (df['perceived_threat'] + df['perceived_competition']) / 2
df.head()
```

```python
summary = (
    df.groupby(['condition', 'institution_context'])
    .agg(
        n=('participant', 'size'),
        mean_stereotype_strength=('stereotype_strength', 'mean'),
        mean_prejudice=('prejudice_rating', 'mean'),
        mean_discrimination=('discrimination_tendency', 'mean'),
        mean_contact_support=('contact_support_index', 'mean'),
        mean_threat_competition=('threat_competition_index', 'mean')
    )
    .reset_index()
)
summary
```

# In-Group Bias in Social Psychology

This notebook scaffold loads the in-group bias dataset and prepares summaries for target relation, identity salience, threat, trust, fairness, resource allocation, moral judgment, punishment, and cooperation.

```python
from pathlib import Path
import pandas as pd

data_path = Path('../data/ingroup_bias_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/ingroup_bias_model.py --simulate first')

df = pd.read_csv(data_path)
df.head()
```

```python
summary = (
    df.groupby(['condition', 'target_group_relation'])
    .agg(
        n=('trust_rating', 'size'),
        mean_trust=('trust_rating', 'mean'),
        mean_fairness=('fairness_rating', 'mean'),
        mean_resource=('resource_allocation', 'mean'),
        mean_blame=('moral_blame', 'mean'),
        mean_punishment=('punishment_severity', 'mean'),
        cooperation_rate=('cooperation_choice', 'mean')
    )
    .reset_index()
)
summary
```

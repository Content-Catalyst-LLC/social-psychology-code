# Intergroup Conflict

This notebook scaffold loads intergroup-conflict data and prepares summaries for resource competition, zero-sum perception, identity salience, status threat, symbolic threat, realistic threat, stereotypes, contact, legitimacy, hostility, exclusion, aggression, and cooperation.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/intergroup_conflict_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/intergroup_conflict_model.py --simulate first')

df = pd.read_csv(data_path)
df['material_conflict_index'] = (df['resource_competition'] + df['zero_sum_perception'] + df['realistic_threat']) / 3
df['identity_threat_index'] = (df['identity_salience'] + df['group_identification'] + df['status_threat'] + df['symbolic_threat']) / 4
df['cooperative_contact_index'] = (df['contact_quality'] + df['equal_status'] + df['common_goal_salience'] + df['institutional_support']) / 4
df.head()
```

```python
summary = (
    df.groupby(['condition', 'context_type'])
    .agg(
        n=('participant', 'size'),
        mean_hostility=('hostility_score', 'mean'),
        mean_aggression=('aggression_intention', 'mean'),
        mean_exclusion=('support_for_exclusion', 'mean'),
        mean_cooperation=('support_for_cooperation', 'mean'),
        mean_contact=('cooperative_contact_index', 'mean')
    )
    .reset_index()
)
summary
```

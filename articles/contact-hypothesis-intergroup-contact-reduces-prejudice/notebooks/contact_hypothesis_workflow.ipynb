# Contact Hypothesis and Intergroup Contact

This notebook scaffold loads the contact hypothesis dataset and prepares summaries for contact quality, Allport conditions, negative contact, intergroup anxiety, empathy, trust, prejudice change, social distance, and inclusive norms.

```python
from pathlib import Path
import pandas as pd

data_path = Path('../data/contact_hypothesis_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/contact_hypothesis_model.py --simulate first')

df = pd.read_csv(data_path)
df['prejudice_change'] = df['prejudice_post'] - df['prejudice_pre']
df.head()
```

```python
summary = (
    df.groupby('condition')
    .agg(
        n=('prejudice_post', 'size'),
        mean_contact_quality=('contact_quality', 'mean'),
        mean_negative_contact=('negative_contact', 'mean'),
        mean_anxiety=('intergroup_anxiety', 'mean'),
        mean_empathy=('empathy', 'mean'),
        mean_trust=('trust', 'mean'),
        mean_prejudice_change=('prejudice_change', 'mean'),
        mean_social_distance=('social_distance', 'mean')
    )
    .reset_index()
)
summary
```

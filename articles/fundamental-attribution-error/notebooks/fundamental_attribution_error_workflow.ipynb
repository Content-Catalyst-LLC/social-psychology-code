# Fundamental Attribution Error

This notebook scaffold loads the attribution dataset and prepares summaries for dispositional attribution, situational attribution, constraint neglect, correspondence inference, blame, punishment, and correction conditions.

```python
from pathlib import Path
import pandas as pd

data_path = Path('../data/fundamental_attribution_error_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/fundamental_attribution_error_model.py --simulate first')

df = pd.read_csv(data_path)
df['fae_score'] = df['dispositional_attribution'] - df['situational_attribution']
df['constraint_neglect'] = df['actual_constraint'] - df['perceived_constraint']
df.head()
```

```python
summary = (
    df.groupby(['condition', 'behavior_valence'])
    .agg(
        n=('fae_score', 'size'),
        mean_dispositional=('dispositional_attribution', 'mean'),
        mean_situational=('situational_attribution', 'mean'),
        mean_fae=('fae_score', 'mean'),
        mean_constraint_neglect=('constraint_neglect', 'mean'),
        mean_blame=('moral_blame', 'mean'),
        mean_punishment=('punishment_recommendation', 'mean')
    )
    .reset_index()
)
summary
```

# Deindividuation

This notebook scaffold loads deindividuation data and prepares summaries for anonymity, self-awareness, accountability, group identity salience, norm valence, norm congruence, prosocial behavior, and antisocial behavior.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/deindividuation_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/deindividuation_model.py --simulate first')

df = pd.read_csv(data_path)
df['identity_shift_index'] = df['group_identity_salience'] - df['personal_identity_salience']
df['deindividuation_index'] = (df['anonymity'] + df['crowd_immersion'] + df['responsibility_diffusion'] + df['arousal_index'] - df['self_awareness'] - df['accountability']) / 4
df.head()
```

```python
summary = (
    df.groupby(['condition', 'context_type'])
    .agg(
        n=('participant', 'size'),
        mean_behavior=('behavior_score', 'mean'),
        mean_prosocial=('prosocial_behavior', 'mean'),
        mean_antisocial=('antisocial_behavior', 'mean'),
        mean_anonymity=('anonymity', 'mean'),
        mean_self_awareness=('self_awareness', 'mean'),
        mean_accountability=('accountability', 'mean'),
        mean_norm_congruence=('norm_congruence', 'mean')
    )
    .reset_index()
)
summary
```

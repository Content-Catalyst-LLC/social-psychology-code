# Obedience, Authority, and Social Power

This notebook scaffold loads obedience-research data and prepares summaries for authority legitimacy, responsibility displacement, moral conflict, peer dissent, obedience, resistance, hesitation, protest, and response time.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/obedience_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/obedience_authority_model.py --simulate first')

df = pd.read_csv(data_path)
df['authority_pressure_index'] = (
    df['authority_legitimacy'] + df['authority_proximity'] + df['institutional_prestige'] +
    df['command_clarity'] + df['cost_of_defiance'] + df['peer_compliance'] - df['peer_dissent']
) / 6
df['moral_resistance_index'] = (
    df['moral_conflict'] + df['victim_proximity'] + df['harm_salience'] +
    df['peer_dissent'] + df['perceived_responsibility_after']
) / 5
df.head()
```

```python
summary = (
    df.groupby(['condition', 'institution_context'])
    .agg(
        n=('participant', 'size'),
        obedience_rate=('obeyed', 'mean'),
        resistance_rate=('resisted', 'mean'),
        protest_rate=('protest', 'mean'),
        mean_authority_pressure=('authority_pressure_index', 'mean'),
        mean_moral_resistance=('moral_resistance_index', 'mean'),
        mean_hesitation=('hesitation', 'mean')
    )
    .reset_index()
)
summary
```

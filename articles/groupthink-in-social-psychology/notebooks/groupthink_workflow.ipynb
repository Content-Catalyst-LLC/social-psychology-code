# Groupthink in Social Psychology

This notebook scaffold loads groupthink data and prepares summaries for cohesion, directive leadership, group insulation, stress, consensus pressure, self-censorship, mindguarding, dissent visibility, outside information, decision safeguards, and decision quality.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/groupthink_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/groupthink_model.py --simulate first')

df = pd.read_csv(data_path)
df['symptom_index'] = df[[
    'illusion_invulnerability', 'collective_rationalization', 'inherent_morality',
    'outgroup_stereotyping', 'self_censorship', 'direct_pressure_dissenters',
    'mindguarding'
]].mean(axis=1)
df['antecedent_risk_index'] = df[[
    'cohesion', 'leadership_directive', 'group_insulation',
    'stress_level', 'consensus_pressure'
]].mean(axis=1)
df['process_quality_index'] = df[[
    'dissent_visibility', 'outside_information', 'alternative_search',
    'risk_analysis', 'contingency_planning', 'leader_impartiality',
    'psychological_safety'
]].mean(axis=1)
df['groupthink_risk_index'] = df['antecedent_risk_index'] + df['symptom_index'] - df['process_quality_index']
df.head()
```

```python
summary = (
    df.groupby(['condition', 'institution_context'])
    .agg(
        n=('participant', 'size'),
        mean_groupthink_risk=('groupthink_risk_index', 'mean'),
        mean_self_censorship=('self_censorship', 'mean'),
        mean_dissent_visibility=('dissent_visibility', 'mean'),
        mean_decision_quality=('decision_quality', 'mean'),
        mean_implementation_risk=('implementation_risk', 'mean')
    )
    .reset_index()
)
summary
```

# Social Norms

This notebook scaffold loads social-norms data and prepares summaries for descriptive norms, injunctive norms, empirical expectations, normative expectations, sanctions, legitimacy, pluralistic ignorance, dynamic norms, compliance, contribution, and norm-message effects.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/social_norms_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/social_norms_model.py --simulate first')

df = pd.read_csv(data_path)
df['norm_strength_index'] = (df['descriptive_norm'] + df['injunctive_norm'] + df['empirical_expectation'] + df['normative_expectation']) / 4
df['tipping_margin'] = df['tipping_exposure'] - df['norm_threshold']
df.head()
```

```python
summary = (
    df.groupby(['condition', 'policy_domain', 'message_type'])
    .agg(
        n=('participant', 'size'),
        compliance_rate=('complied', 'mean'),
        mean_intention=('compliance_intention', 'mean'),
        mean_contribution=('contribution_amount', 'mean'),
        mean_descriptive=('descriptive_norm', 'mean'),
        mean_injunctive=('injunctive_norm', 'mean'),
        mean_legitimacy=('institutional_legitimacy', 'mean')
    )
    .reset_index()
)
summary
```

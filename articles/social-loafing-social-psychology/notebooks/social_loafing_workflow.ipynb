# Social Loafing Workflow
Generate data with `python/social_loafing_model.py`, then load and summarize effort loss.

```python
from pathlib import Path
import pandas as pd
import numpy as np

data_path = Path('../data/social_loafing_trials.csv')
df = pd.read_csv(data_path)
df['accountability_index'] = (df['identifiability'] + df['accountability'] + df['task_visibility'] + df['evaluation_potential'] + df['digital_traceability']) / 5
df.head()
```

```python
df.groupby(['condition','task_type']).agg(n=('participant','size'), mean_effort_loss=('effort_loss','mean'), mean_motivation_loss=('motivation_loss','mean'), mean_accountability=('accountability_index','mean')).reset_index()
```

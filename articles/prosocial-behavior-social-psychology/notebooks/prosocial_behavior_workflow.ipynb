# Prosocial Behavior

This notebook scaffold loads prosocial-behavior data and prepares summaries for helping, donation, volunteering, cooperation, empathy, norms, efficacy, cost, responsibility, identity, trust, and institutional legitimacy.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/prosocial_behavior_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/prosocial_behavior_model.py --simulate first')

df = pd.read_csv(data_path)
df['log_bystanders'] = np.log1p(df['bystander_count'])
df['prosocial_motivation_index'] = (df['empathy_score'] + df['perspective_taking'] + df['norm_salience'] + df['efficacy_belief'] + df['felt_responsibility'] + df['moral_identity'] - df['helping_cost'] - df['intervention_risk']) / 6
df.head()
```

```python
summary = (
    df.groupby(['condition', 'context_type'])
    .agg(
        n=('participant', 'size'),
        helping_rate=('helping_decision', 'mean'),
        mean_donation=('donation_amount', 'mean'),
        mean_cooperation=('cooperation_contribution', 'mean'),
        mean_empathy=('empathy_score', 'mean'),
        mean_norms=('norm_salience', 'mean'),
        mean_efficacy=('efficacy_belief', 'mean')
    )
    .reset_index()
)
summary
```

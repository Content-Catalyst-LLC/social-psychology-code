# Self-Serving Bias in Social Psychology

This notebook scaffold loads the self-serving bias dataset and prepares summaries for success/failure attribution, internal/external attribution, credit claiming, excuse making, accountability, and learning intention.

```python
from pathlib import Path
import pandas as pd

data_path = Path('../data/self_serving_bias_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/self_serving_bias_model.py --simulate first')

df = pd.read_csv(data_path)
df.head()
```

```python
summary = (
    df.groupby(['condition', 'outcome_valence', 'self_other'])
    .agg(
        n=('internal_attribution', 'size'),
        mean_internal=('internal_attribution', 'mean'),
        mean_external=('external_attribution', 'mean'),
        mean_responsibility=('responsibility_rating', 'mean'),
        mean_credit=('credit_claiming', 'mean'),
        mean_excuse=('excuse_making', 'mean'),
        mean_learning=('learning_intention', 'mean')
    )
    .reset_index()
)
summary
```

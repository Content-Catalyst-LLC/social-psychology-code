# Social Dilemmas

This notebook scaffold loads social-dilemma data and prepares summaries for contributions, extraction, trust, norms, enforcement, legitimacy, resource stock, and welfare.

```python
from pathlib import Path
import pandas as pd

data_path = Path('../data/social_dilemmas_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/social_dilemmas_model.py --simulate first')

df = pd.read_csv(data_path)
df['free_riding_index'] = (df['endowment'] - df['contribution']) / df['endowment']
df.head()
```

```python
summary = (
    df.groupby(['condition', 'dilemma_type'])
    .agg(
        n=('participant', 'size'),
        mean_contribution=('contribution', 'mean'),
        mean_extraction=('extraction', 'mean'),
        mean_trust=('trust_score', 'mean'),
        mean_norm_salience=('norm_salience', 'mean'),
        mean_legitimacy=('institutional_legitimacy', 'mean'),
        mean_group_welfare=('group_welfare', 'mean')
    )
    .reset_index()
)
summary
```

# Tragedy of the Commons

This notebook scaffold loads commons-extraction data and prepares summaries for resource stock, extraction, trust, legitimacy, monitoring, boundaries, rule participation, stewardship norms, and depletion risk.

```python
from pathlib import Path
import pandas as pd

data_path = Path('../data/tragedy_commons_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/tragedy_commons_model.py --simulate first')

df = pd.read_csv(data_path)
df['expected_sanction'] = df['monitoring_credibility'] * df['sanction_probability'] * df['sanction_severity'] / 10
df.head()
```

```python
summary = (
    df.groupby(['condition', 'property_regime'])
    .agg(
        n=('participant', 'size'),
        mean_extraction=('extraction', 'mean'),
        mean_stock=('resource_stock', 'mean'),
        mean_trust=('trust_score', 'mean'),
        mean_legitimacy=('legitimacy_score', 'mean'),
        mean_monitoring=('monitoring_credibility', 'mean'),
        mean_depletion_risk=('depletion_risk', 'mean')
    )
    .reset_index()
)
summary
```

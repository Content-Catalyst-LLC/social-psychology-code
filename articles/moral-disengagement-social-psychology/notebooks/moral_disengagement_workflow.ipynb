# Moral Disengagement in Social Psychology

This notebook scaffold loads the moral disengagement dataset and prepares summaries for mechanism scores, harmful decision rates, policy endorsement, unethical intention, empathy, guilt, and institutional pressure.

```python
from pathlib import Path
import pandas as pd

data_path = Path('../data/moral_disengagement_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/moral_disengagement_model.py --simulate first')

df = pd.read_csv(data_path)
mechanisms = [
    'moral_justification', 'euphemistic_labeling', 'advantageous_comparison',
    'displaced_responsibility', 'diffused_responsibility', 'consequence_distortion',
    'dehumanization', 'blame_attribution'
]
df['md_index'] = df[mechanisms].mean(axis=1)
df.head()
```

```python
summary = (
    df.groupby(['condition', 'scenario_domain'])
    .agg(
        n=('md_index', 'size'),
        mean_md=('md_index', 'mean'),
        harmful_rate=('harmful_decision', 'mean'),
        mean_policy_endorsement=('policy_endorsement', 'mean'),
        mean_unethical_intention=('unethical_intention', 'mean'),
        mean_empathy=('empathy', 'mean'),
        mean_guilt=('guilt', 'mean')
    )
    .reset_index()
)
summary
```

# Implicit Bias in Social Psychology

This notebook scaffold loads IAT-style trial data, computes D-score-style summaries, and prepares analyses for response latency, explicit attitudes, judgment, behavioral outcomes, and institutional mitigation.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/implicit_bias_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/implicit_bias_model.py --simulate first')

df = pd.read_csv(data_path)
df['log_response_time'] = np.log(df['response_time_ms'])
df.head()
```

```python
valid = df[(df['response_time_ms'] >= 250) & (df['accuracy'] == 1)].copy()
means = (
    valid.groupby(['participant', 'condition', 'congruent_block'], observed=True)['response_time_ms']
    .mean()
    .reset_index()
)
wide = means.pivot_table(index=['participant', 'condition'], columns='congruent_block', values='response_time_ms').reset_index()
wide.columns = ['participant', 'condition', 'mean_incongruent', 'mean_congruent']
wide['simple_latency_difference'] = wide['mean_incongruent'] - wide['mean_congruent']
wide.head()
```

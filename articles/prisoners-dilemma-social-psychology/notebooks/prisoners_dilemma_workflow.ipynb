# Prisoner's Dilemma

This notebook scaffold loads the prisoner’s dilemma dataset and prepares summaries for cooperation, payoff, trust, fairness, reciprocity, reputation, institutional enforcement, and repeated-game dynamics.

```python
from pathlib import Path
import pandas as pd

data_path = Path('../data/prisoners_dilemma_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/prisoners_dilemma_model.py --simulate first')

df = pd.read_csv(data_path)
df['temptation_gap'] = df['temptation_payoff'] - df['reward_payoff']
df.head()
```

```python
summary = (
    df.groupby(['condition', 'round'])
    .agg(
        n=('cooperate', 'size'),
        cooperation_rate=('cooperate', 'mean'),
        mean_payoff=('own_payoff', 'mean'),
        mean_trust=('trust_score', 'mean'),
        mean_fairness=('fairness_score', 'mean'),
        mean_expected_partner_cooperation=('expected_partner_cooperation', 'mean')
    )
    .reset_index()
)
summary
```

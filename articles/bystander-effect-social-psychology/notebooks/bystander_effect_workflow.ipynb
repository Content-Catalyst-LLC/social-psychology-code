# The Bystander Effect

This notebook scaffold loads bystander-effect data and prepares summaries for perceived bystander count, emergency clarity, diffusion of responsibility, evaluation apprehension, direct assignment, intervention likelihood, actual intervention, and intervention latency.

```python
from pathlib import Path
import numpy as np
import pandas as pd

data_path = Path('../data/bystander_effect_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/bystander_effect_model.py --simulate first')

df = pd.read_csv(data_path)
df['log_perceived_bystanders'] = np.log1p(df['perceived_bystander_count'])
df['ambiguity_index'] = (df['pluralistic_ignorance'] + df['evaluation_apprehension'] + (10 - df['emergency_clarity'])) / 3
df.head()
```

```python
summary = (
    df.groupby(['condition', 'context_type'])
    .agg(
        n=('participant', 'size'),
        intervention_rate=('actual_intervention', 'mean'),
        mean_likelihood=('intervention_likelihood', 'mean'),
        mean_latency=('intervention_latency_ms', 'mean'),
        mean_bystanders=('perceived_bystander_count', 'mean'),
        mean_responsibility=('felt_responsibility', 'mean'),
        mean_diffusion=('diffusion_responsibility', 'mean')
    )
    .reset_index()
)
summary
```

# Diffusion of Responsibility

This notebook scaffold loads the diffusion-of-responsibility dataset and prepares summaries for bystander count, role clarity, ambiguity, perceived responsibility, intervention decisions, reporting decisions, and intervention delay.

```python
from pathlib import Path
import pandas as pd

data_path = Path('../data/diffusion_responsibility_trials.csv')
if not data_path.exists():
    raise FileNotFoundError('Run python/diffusion_responsibility_model.py --simulate first')

df = pd.read_csv(data_path)
df['pluralistic_ignorance_gap'] = df['private_concern'] - df['perceived_group_concern']
df.head()
```

```python
summary = (
    df.groupby(['condition', 'scenario_domain'])
    .agg(
        n=('intervention_decision', 'size'),
        mean_bystanders=('bystander_count', 'mean'),
        intervention_rate=('intervention_decision', 'mean'),
        reporting_rate=('reporting_decision', 'mean'),
        mean_responsibility=('perceived_responsibility', 'mean'),
        mean_role_clarity=('role_clarity', 'mean'),
        mean_ambiguity=('ambiguity_level', 'mean'),
        mean_delay=('intervention_delay_seconds', 'mean')
    )
    .reset_index()
)
summary
```
